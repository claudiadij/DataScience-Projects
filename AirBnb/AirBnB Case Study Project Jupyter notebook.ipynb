```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available?
print(f'Total Neighborhood Groups: {len(air_bnb.neighbourhood_group.unique())}')

# Which shows up the most?
boroughs = air_bnb.groupby('neighbourhood_group')['id'].count()
print(boroughs)

print('Manhattan is the neighborbood that shows up the most.')
```

```python
# Are private rooms the most popular in manhattan?
manhattan = air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan']
manhattan

print(manhattan.groupby('room_type').sum()['number_of_reviews'])
print('Entire home/apt are the most popular stays in Manhattan based on the total number of reviews.')
```

```python
# Which hosts are the busiest and based on their reviews?
review_sums = air_bnb.groupby('host_id').sum()['number_of_reviews']
print(review_sums.sort_values(ascending = False))

print(air_bnb[air_bnb['host_id']==37312959])
print(air_bnb[air_bnb['host_id']==344035])
print(air_bnb[air_bnb['host_id']==26432133])
print(air_bnb[air_bnb['host_id']==35524316])
print(air_bnb[air_bnb['host_id']==40176101])

print('The top 5 busiest hosts are: Maya (host_id: 37312959), Brooklyn & Breakfast (host_id: 344035), Danielle(host_id: 26432133), Yasu & Akiko(host_id: 35524316), and Brady (host_id: 40176101)')
```

```python
# Which neighorhood group has the highest average price?
print(air_bnb.groupby('neighbourhood_group')['price'].mean())

print('Manhattan has the heighest average of $196.88')
```

```python
# Which neighbor hood group has the highest total price?
print(air_bnb.groupby('neighbourhood_group')['price'].sum())

print('Manhattan has the highest total price.')
```

```python
#Which top 5 hosts have the highest total price?
highest_total = air_bnb.groupby('host_name').sum().sort_values('price', ascending = False)
print(highest_total.head(5))

print('Top 5 Hosts: Sonder(NYC) (host_id: 71782340547), Blueground (host_id: 24924786136), Michael (host_id: 22673153604), David (host_id: 28633613354), Alex (host_id: 19850092761)')
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
no_availability = len(air_bnb[(air_bnb['availability_365'] == 0) & (air_bnb['number_of_reviews'] >= 100)])
no_availability

print(f'There are {no_availability} listings with no availability with a review of 100+')
```

```python
# What host has the highest total of prices and where are they located??
print(air_bnb.groupby('host_id').sum()[['id','price']].sort_values('price', ascending = False))
print(air_bnb[air_bnb['host_id']==219517861])

print('Sonder has the highest total of prices and is located in the Financial District')
```

```python
# When did Danielle from Queens last receive a review?
print(air_bnb[(air_bnb['host_name']=='Danielle') & (air_bnb['neighbourhood_group']=='Queens')]['last_review'])

print("Danielle's last review was on 2019-07-08")
```
