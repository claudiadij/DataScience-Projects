```python
import pandas as pd
import numpy as np

allstar = pd.read_csv('./NBA All Stars 2000-2016 - Sheet1.csv')
allstar
```

# What is the average weight of all players who played during this time?

```python
allstar.describe(include = 'all')['WT']['mean']

print('The average weight of all players who played during this time was: 228.76')
```

# What team is represented the most during this time?

```python
allstar.groupby('Team').count().sort_values('Year',ascending=False)[['Year']].head()

print('Miami Heat was represented the most during this time.')
```

# What draft class is represeted the most during this time?

```python
allstar.groupby('NBA Draft Status').count().sort_values('Player',ascending=False)[['Player']].head()

print('1996 Rnd 1 Pick 13 was represented the most during this time.')
```

# Are foriegn players more prevalent in the All-Star Game during this time?

```python
allstar.groupby('Nationality').count().sort_values('Player', ascending = False)[['Player']].head()

print('Foreign players are not more prevalent than American players.')
```

# How often are the Western All-Stars voted in by fans?

```python
allstar[allstar['Selection Type']=='Western All-Star Fan Vote Selection'].groupby('Year').count()[['Player']]

print('The Western All-Stars were voted in by fans every year.')
```

# How many times does Steph Curry make the All-Star game during this time period?

```python
allstar.groupby('Player').count().sort_values('Player', ascending = False)

print('Steph Curry makes the All-Star game four times.')
```

# How many Shooting Gaurds have made the All-Star game during this time?

```python
allstar[allstar['Pos'] == 'SG'].count() # ans: 57

# Drop duplicates
allstar_nodups = allstar.drop_duplicates(['Player'])
allstar_nodups[allstar['Pos'] == 'SG'].count()

print('Before dropping duplicate players, 57 Shooting Guards have made the All-Star game, but after dropping duplicates, only 9 Shooting Guards have made the All-Star game during this time.')
```

# How many Chicago Bulls players have made the All-Star Game during this time period?

```python
allstar.groupby(['Team']).count() # ans: 12

# Drop duplicates
allstar_nodups.groupby(['Team']).count()

print('Before dropping duplicate players, 12 Chicago Bulls players have made the All-Star game, but after dropping duplicates, only 5 Chicago Bulls players have made the All-Star game during this time.')
```
